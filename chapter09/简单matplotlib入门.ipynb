# 1.简明入门

```python
import matplotlib.pyplot as plt
import numpy as np
%config IPCompleter.greedy=True

#必须加入如下代码，否则会看不到图形输出
%matplotlib inline

data=np.arange(10)
plt.plot(data)
```

# 2.使用add_subplot方法一次绘制多个图

使用add_subplot方法绘制2行2列共3个图

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)#第1张图
ax1.plot(np.random.randn(50).cumsum(),'r--')
ax2=fig.add_subplot(2,2,2)#第2张图
ax2.plot(np.random.randn(50).cumsum(),'g--')
ax3=fig.add_subplot(2,2,3)#第3张图
ax3.plot(np.random.randn(50).cumsum(),'b--')
```

# 3.使用subplots方法一次绘制多个图

除了使用add_subplot方法绘制多图外，我们也可以使用subplots方法绘制多图，该方法返回的是多个AxesSubplot对象就是上面的ax1,ax2等对象。

```python
fig,axes=plt.subplots(2,2)
axes[0][0].plot(np.random.randn(50).cumsum(),'b--')
```
